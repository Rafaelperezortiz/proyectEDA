```python
import pandas as pd

test_data = pd.read_csv("/workspaces/machine-learning-python-template/AB_NYC_2019.csv")


train_data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv")
total_data = pd.concat([train_data,test_data]).reset_index(inplace=False, drop = True)
total_data.head()
```

```python
total_data.drop_duplicates()
```

```python
#sacar dimensiones del dataframe .shape
total_data.shape
```

```python
total_data.info()
```

```python
total_data.drop(["reviews_per_month","id","host_id","name","latitude","longitude","host_name","last_review"], axis = 1, inplace= True)
total_data.info()
```

```python
total_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(2, 3, figsize = (10,7))

sns.histplot(data = total_data, x = "price",  ax = axis[0,0])
sns.boxenplot(data = total_data, x = "minimum_nights",  ax = axis[0,1])
sns.histplot(data = total_data, x = "neighbourhood_group",  ax = axis[0,2])
sns.histplot(data = total_data, x = "room_type",  ax = axis[1,0])
sns.histplot(data = total_data, x = "number_of_reviews",  ax = axis[1,1])
sns.histplot(data = total_data, x = "availability_365",  ax = axis[1,2])

plt.tight_layout()

plt.show()
```

```python
import seaborn as sns

fig, axis = plt.subplots(2, 2, figsize = (40,10))

sns.boxplot(data = total_data, x = "price",  ax = axis[0,0])
sns.boxplot(data = total_data, x = "neighbourhood",  ax = axis[0,1])
sns.boxplot(data = total_data, x = "minimum_nights",  ax = axis[1,0])
sns.boxplot(data = total_data, x = "number_of_reviews",  ax = axis[1,1])


plt.tight_layout()

plt.show()
```

```python
total_data["room_type"] = pd.factorize(total_data['room_type'])[0]
total_data.head()
```

```python
#total_data["neighbourhood_group"] = pd.factorize("neighbourhood_group")
total_data["neighbourhood_group"] = pd.factorize(total_data["neighbourhood_group"])[0]
total_data.head()
```

```python
total_data["neighbourhood"] = pd.factorize(total_data["neighbourhood"])[0]
total_data.head()
```

```python
fig, axes = plt.subplots(2,3, figsize =(13,8))

sns.barplot(data = total_data, x = "neighbourhood_group", y = "price", ax = axes[0,0])
sns.heatmap(total_data[[ "price", "neighbourhood_group"]].corr(), annot= True, fmt= ".2f", ax = axes[1,0])
sns.barplot(data = total_data, x = "neighbourhood", y ="price", ax = axes[0,1])
sns.heatmap(total_data[[ "price", "neighbourhood"]].corr(), annot= True, fmt= ".2f", ax = axes[1,1])           
sns.barplot(data = total_data, x = "room_type", y = "price", ax = axes[0,2])
sns.heatmap(total_data[[ "price", "room_type"]].corr(), annot= True, fmt= ".2f", ax = axes[1,2])


plt.tight_layout()
plt.show()
```
